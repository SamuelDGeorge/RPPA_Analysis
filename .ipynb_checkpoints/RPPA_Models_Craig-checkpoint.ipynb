# Prepare Data for Machine Learning

This work flow will break apart a training set and test set as well as put the data through a pipeline to prepare it for fitting to a model

```python
split = StratifiedShuffleSplit(n_splits=1, test_size=0.3, random_state=42)

#For all 3 conditions
for train_index, test_index in split.split(raw_data, raw_data['Cell line']):
    strat_train = raw_data.loc[train_index]
    strat_test = raw_data.loc[test_index]  

#DMSO v P600
for train_index, test_index in split.split(raw_data_P600, raw_data_P600['Cell line']):
    strat_train_P600 = raw_data_P600.loc[train_index]
    strat_test_P600 = raw_data_P600.loc[test_index]
    
#P600 v P600W    
for train_index, test_index in split.split(raw_data_P600W, raw_data_P600W['Cell line']):
    strat_train_P600W = raw_data_P600W.loc[train_index]
    strat_test_P600W = raw_data_P600W.loc[test_index]
    
#DMSO v P600W    
for train_index, test_index in split.split(raw_data_DMSO_P600W, raw_data_DMSO_P600W['Cell line']):
    strat_train_DMSO_P600W = raw_data_DMSO_P600W.loc[train_index]
    strat_test_DMSO_P600W = raw_data_DMSO_P600W.loc[test_index]
    
strat_train
```

```python
#Cell Line
raw_train_cell = strat_train.drop('Cell line', axis=1)
train_labels_cell = pd.DataFrame(strat_train['Cell line'].copy())

raw_test_cell = strat_test.drop('Cell line', axis=1)
test_labels_cell = pd.DataFrame(strat_test['Cell line'].copy())

#All 3 conditions
raw_train_all = strat_train.drop('Drug', axis=1)
train_labels_all = pd.DataFrame(strat_train['Drug'].copy())

raw_test_all = strat_test.drop('Drug', axis=1)
test_labels_all = pd.DataFrame(strat_test['Drug'].copy())

#DMSO vs P600
raw_train_P600 = strat_train_P600.drop('Drug', axis=1)
train_labels_P600 = pd.DataFrame(strat_train_P600['Drug'].copy())

raw_test_P600 = strat_test_P600.drop('Drug', axis=1)
test_labels_P600 = pd.DataFrame(strat_test_P600['Drug'].copy())

#P600 vs P600W
raw_train_P600W = strat_train_P600W.drop('Drug', axis=1)
train_labels_P600W = pd.DataFrame(strat_train_P600W['Drug'].copy())

raw_test_P600W = strat_test_P600W.drop('Drug', axis=1)
test_labels_P600W = pd.DataFrame(strat_test_P600W['Drug'].copy())

#DMSO vs P600W
raw_train_DMSO_P600W = strat_train_DMSO_P600W.drop('Drug', axis=1)
train_labels_DMSO_P600W = pd.DataFrame(strat_train_DMSO_P600W['Drug'].copy())

raw_test_DMSO_P600W = strat_test_DMSO_P600W.drop('Drug', axis=1)
test_labels_DMSO_P600W = pd.DataFrame(strat_test_DMSO_P600W['Drug'].copy())


raw_train_all
```

# Ready for Machine Learning:

At this point the data is in the proper format for training your model. Relevant variables listed below:

The Training Data:

x_train_data - The X in your model. These are what we are going to use to make predictions. 

y_train_data - The Y in your model. These are the values you are going to try and predict.

The Test Data:

x_test_data - The X data for your test data. Use this to test your completed model 

y_test_data - the Y data for your test data. Use this to test your model againt real predictions

# Random Forest Model with RandomSearchCV


Uses a random forest and GridSearchCV to predict the hyperparameters and generate a model

```python
X_train = train_cell
Y_train = train_labels_cell

X_test = test_cell
Y_test = test_labels_cell
```

```python
#Do a random search for best iterations

param_grid_random = {
        'n_estimators': list(range(300,400)),
        'max_features': list(range(30,150)),
    }

forest_reg_rand_clf = RandomForestClassifier(bootstrap = True)

for_random_clf = RandomizedSearchCV(forest_reg_rand_clf, param_distributions=param_grid_random,
                                n_iter=1000, cv=5, scoring='f1_macro',
                                verbose=2, n_jobs=-1, random_state=42)


for_random_clf.fit(X_train, Y_train)
```
